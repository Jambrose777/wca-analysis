# World Cube Association - Forecasting

**Speedcubing** is the activity of solving a variety of twisty puzzles as fast as possible. The most famous twisty puzzle from the Speedcubing domain (which is also responsible for shaping the term *Speedcubing*) is the widely-known [Rubik's Cube](https://en.wikipedia.org/wiki/Rubik%27s_Cube), or more generally **3x3x3 Cube** (since *Rubik's Cube* is technically a brand name). 

In recent years, the popularity of Speedcubing has grown exponentially all over the world, which can mainly be attributed to the efforts of its global governing body, the **[World Cube Association (WCA)](https://www.worldcubeassociation.org/)**. In 2017 alone, there have been 925 WCA competitions in 88 different countries with 37547 distinct competitors.

In this notebook, I will attempt to make forecasts for a few WCA statistics for the near future, based the public [WCA database export](https://www.worldcubeassociation.org/results/misc/export.html).

For now, I will limit this project to forcasting competitions per month and participants per month, but I might extend it in the future. :)

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import statsmodels.api as sm
import datetime
```

## Import and prepare data

For my analysis, I will use 2018-03-31's WCA export. First, I import the results table from the WCA export. This gives us a very large dataframe with all results from official WCA competitions. Right away, I will only import the columns that will be needed subsequently. 

```python
results = pd.read_csv('db_export/WCA_export_Results.tsv', delimiter='\t',
            usecols=['competitionId', 'eventId', 'personId', 'personCountryId'] + ['value' + str(i+1) for i in range(5)])
results.head()
```

```python
results.info()
```

Since for this analysis, I am not interested in individual results, I first reduce the size of this dataframe by transforming the individual results into number of solves per round, then I group them per competition, event and person.

```python
results['solves'] = sum((results['value' + str(i+1)] > 0) for i in range(5))
df = results.drop(columns=['value' + str(i+1) for i in range(5)])
df = df.groupby(['competitionId', 'eventId', 'personId', 'personCountryId'], as_index=False).sum()
df.info()
```

With this being done, I can now fetch more data and extend the current dataframe as follows:

1. I will use the start date of each competition so assign a timestamp to each row (which will then be used for a DateTimeIndex)
2. I will also add a continentId for each row for possible per continent/country analyses

```python
# Fetching competition data
competition_data = pd.read_csv('db_export/WCA_export_Competitions.tsv', delimiter='\t', 
                               usecols=['id','countryId','year','month','day'])

def get_start_date(row):
    year, month, day = row['year'], row['month'], row['day']
    return pd.to_datetime(datetime.date(year,month,day).strftime('%Y-%m-%d'))

competition_data['date'] = pd.to_datetime(competition_data.apply(get_start_date, axis=1))

df2 = competition_data[['date','id','countryId']].merge(df, how='inner', left_on='id', right_on='competitionId')
df3 = df2.drop(columns=['id'])

countries = pd.read_csv('db_export/WCA_export_Countries.tsv', delimiter='\t', usecols=['id', 'continentId'])
countries['continent'] = countries['continentId'].str[1:]

df4 = df3.merge(countries[['id', 'continent']], how='inner', left_on='countryId', right_on='id').drop(columns=['id'])
df4 = df4.rename(columns={'countryId': 'compCountryId', 'continent': 'compContinentId'})
df5 = df4.merge(countries[['id', 'continent']], how='inner', left_on='personCountryId', right_on='id').drop(columns=['id'])
df5 = df5.rename(columns={'continent': 'personContinentId'})
wca_data = df5.sort_values('date').set_index('date')
wca_data.head()
```

## Competions per month

```python
comps = wca_data['2003':'2018-02-28'].resample('M').competitionId.nunique()
comps.plot()
plt.show()
```

Looks like the WCA is growing consistently! :)

What I plan to do now is a so-called time series forecasting using a SARIMA model. To build such a forecasting model for a time series, it is important to transform the series into a (near-)stationary one, i.e. with consistent mean and standard variation. This process might include several steps. I will start by preparing a function that will make it easier later on to perform repeated stationarity checks.

```python
from statsmodels.tsa.stattools import adfuller

def validate_stationarity(series):
    """ This function plots a time series including rolling mean and standard deviation. 
    It also performs the Dickey-Fuller Test, a statistical test to check for stationarity. """
    
    plt.figure(figsize=(12,6))
    plt.plot(series, label="Data")
    moving_avg = series.rolling(window=12).mean()
    plt.plot(moving_avg, label='Rolling mean')
    moving_sd = series.rolling(window=12).std()
    plt.plot(moving_sd, label='Rolling SD')
    plt.legend(loc='best')
    plt.title('Data with Rolling Mean & Standard Deviation', fontsize=14)
    plt.show()

    dftest = adfuller(series, autolag='AIC')
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

Let's do a first check:

```python
validate_stationarity(comps)
```

Clearly not stationary. What can be seen above is the following:

1. The trend does not increase linearly and the standard deviation is not near-constant. I will resolve this by applying a square root transformation.
2. The mean is not near-constant. This requires so-called differencing.

Applying both gives:

```python
comps_sqrt = np.sqrt(comps)
comps_sqrt_1diff = comps_sqrt.diff(periods=1)
comps_sqrt_2diff = comps_sqrt_1diff.diff(periods=1)
comps_sqrt_2diff_sdiff = comps_sqrt_2diff.diff(periods=3)
validate_stationarity(comps_sqrt_2diff_sdiff.dropna())
```

And now we already have a stationary time series (p << 0.01)! 

Next, I want to find good parameters for a SARIMA forcasting model. For anyone interested, detailed information about the SARIMA model and its parameters can be found [here](https://www.otexts.org/fpp/8/9).

Finding good SARIMA parameters can be done by looking at the autocorrelation function (ACF) and the partial autocorrelation function (PACF). However, doing this preceisely is not that easy. A good compromise is getting a nseful parameter range from looking at ACF/PACF and then finding the best parameters from this range via grid search. I will do this in the following:

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(comps_sqrt_2diff_sdiff.dropna(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(comps_sqrt_2diff_sdiff.dropna(), lags=40, ax=ax2)
```

```python
from itertools import product
warnings.filterwarnings('ignore')

def auto_sarima(train, p=range(3), d=[1], q=range(3), P=range(3), D=[1], Q=range(3), S=12): 
    """ function to perform SARIMA grid search """
    
    pdq = list(product(p, d, q))
    seasonal_pdq = list(product(P, D, Q, [S]))
    aic_scores = []

    for i in pdq:
        for j in seasonal_pdq:
            try:
                model = sm.tsa.statespace.SARIMAX(train, order=i, seasonal_order=j)
                results = model.fit()
                if np.isnan(results.aic) == False:
                    aic_scores.append((i, j, results.aic))
            except: # some parameter combinations will not be valid
                pass
            
    return aic_scores
```

```python
scores = auto_sarima(comps_sqrt, p=range(8), d=[2], q=range(8), S=3)
scores.sort(key=lambda x: x[2])
print('Top 5 parameters (according to AIC scores):')
for t in scores[:5]:
    print('SARIMA{}{} - AIC: {}'.format(*t))
```

Ok, let's build a model using the best parameters from the grid search:

```python
model_competitions = sm.tsa.statespace.SARIMAX(comps_sqrt, order=(7, 2, 0), seasonal_order=(2, 1, 1, 3)).fit()
model_competitions.summary()
```

Great! Especially, all p-values are below 0.001, which indicates a very good parameter choice. 

In general, it is also important to look at the model'S diagnostics to make sure that there is no unusual behavior. The most important aspect is generally, that the error terms (residuals) are (near-)normally distributed. For my model, everything looks very good:

```python
model_competitions.plot_diagnostics(figsize=(15, 12))
plt.show()
```

Finally, here are some relevant error metrics for the model applied to the existing data:

```python
from sklearn.metrics import mean_absolute_error, explained_variance_score

pred_sqrt = model_competitions.predict(start=0, end=comps.index.get_loc(max(comps.index)))
pred = np.square(pred_sqrt)
print('Mean error:', (pred - comps).mean())
print('Mean absolute error:', mean_absolute_error(comps, pred))
print('Explained variance score:', explained_variance_score(comps, pred))
```

**Time for forecasting!**

```python
pred_sqrt = model_competitions.get_forecast(steps=24)
pred_ci_sqrt = pred_sqrt.conf_int(alpha=0.05)
pred = np.square(pred_sqrt.predicted_mean)
pred_ci = np.square(pred_ci_sqrt)

ax = comps['2014':].plot(style='.--', label='Observed', figsize=(16, 12), linewidth=1)
pred.plot(style='.--', ax=ax, label='Forecast', linewidth=1)
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.1, label='Confidence interval')

ax.set_xlabel('')
ax.set_ylabel('Number of WCA competitions', fontsize=20)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(14)

plt.legend(loc='upper left', fontsize=20)
plt.show()
```

Looks pretty cool! Let's hope that this prediction holds for the WCA, or that the growth even exceeds my model's predictions!

## Competitors per month

Now, I will look at the number of people participating in WCA competitions each month. For this sake, people going to multiple competitions in one month will also be counted multiple times.

```python
df = wca_data[['competitionId', 'personId']].drop_duplicates()
df2 = df.reset_index().groupby(['date','competitionId'], as_index=False).count()
df3 = df2.set_index('date')
competitors = df3['2003':'2018-02-28'].resample('M').personId.sum()
validate_stationarity(competitors)
```

The approach for building a forecasting model will generally remain the same as in the previous part, so I will not elaborate much for already seen things.

I will use a more general transformation approach this time, the so-called [Box-Cox-Transformation](https://de.wikipedia.org/wiki/Box-Cox-Transformation). 

However, for using this transformation, the factthat some months in the early years did not have a single WCA competition poses a practical problem. For this reason, I will only work with the data starting from March 2006, since every single month since then had at least one official WCA competition. 

```python
from scipy.stats import boxcox
from scipy.special import inv_boxcox

competitors = competitors['2006-03':]
# Transformation
ct, cl = boxcox(competitors)
competitors_t = competitors.apply(lambda x: boxcox(x, cl))
# First order and seasonal differencing
competitors_t_1diff = competitors_t.diff()
competitors_t_1diff_sdiff = competitors_t_1diff.diff(periods=3)
validate_stationarity(competitors_t_1diff_sdiff.dropna())
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(competitors_t_1diff_sdiff.dropna(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(competitors_t_1diff_sdiff.dropna(), lags=40, ax=ax2)
```

```python
scores = auto_sarima(competitors_t, p=range(7), S=3)
scores.sort(key=lambda x: x[2])
print('Top 5 parameters (according to AIC scores):')
for t in scores[:5]:
    print('SARIMA{}{} - AIC: {}'.format(*t))
```

This time, I will use the 4th best parameter choice, whose parameters are a subset of the top three. I tried the top 3 parameters, and the p-values of the additional terms were not convincing. In such cases, using the less complex model that performs almost equally well is often the better choice.

```python
model_competitors = sm.tsa.statespace.SARIMAX(competitors_t, order=(6, 1, 0), seasonal_order=(0, 1, 2, 3)).fit()
model_competitors.summary()
```

```python
model_competitors.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred_bc = model_competitors.predict(start=0, end=competitors.index.get_loc(max(competitors.index)))
pred = inv_boxcox(pred_bc, cl)
print('Mean error:', (pred - competitors).mean())
print('Mean absolute error:', mean_absolute_error(competitors, pred))
print('Explained variance score:', explained_variance_score(competitors, pred))
```

**Once more: forcasting time!**

```python
pred_bc = model_competitors.get_forecast(steps=24)
pred_ci_bc = pred_bc.conf_int(alpha=0.05)
pred = inv_boxcox(pred_bc.predicted_mean, cl)
pred_ci = inv_boxcox(pred_ci_bc, cl)

ax = competitors['2014':].plot(style='.--', label='Observed', figsize=(16, 12), linewidth=1)
pred.plot(style='.--', ax=ax, label='Forecast', linewidth=1)
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.1, label='Confidence interval')

ax.set_xlabel('')
ax.set_ylabel('Number of people competing in WCA competitions', fontsize=20)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(14)

plt.legend(loc='upper left', fontsize=20)
plt.show()
```
